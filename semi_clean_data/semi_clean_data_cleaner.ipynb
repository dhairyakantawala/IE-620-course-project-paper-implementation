```python
from pathlib import Path
import pandas as pd
from tqdm import tqdm
```

```python
directory_path = Path("semi_clean_data")
files_list = [entry.name for entry in directory_path.iterdir() if entry.is_file()]
```

```python
file_name = 'HDFCB IN Equity.csv'
```

```python
file_name in files_list
```

```python
data = pd.read_csv(f'semi_clean_data/{file_name}', index_col=0)
```

```python
data = data.ffill().dropna(axis=0)
```

```python
for file_name in tqdm(files_list):
    data = pd.read_csv(f'semi_clean_data/{file_name}', index_col=0)
    data = data.ffill().dropna(axis=0)
    data.to_csv(f'final_clean_data/{file_name}')
```
