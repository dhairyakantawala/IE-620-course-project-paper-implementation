```python
from helper_functions import files_list, get_months
months = get_months()
import pandas as pd
import numpy as np
```

```python
features = [
    'ret_1m',
    'ret_3m',
    'mom_12m',
    'vol_3m',
    'vol_6m',
    'vol_z',
    'vol_trend',
    'ey_z',
    'bm_z',
    'mc_z',
    'rev_6m',
    'mom_vol_adj',
    'value_mom',
    'liquidity_adj_mom'
]
```

```python
i = 100
state_i = []
reward_i = []
month = months[i]
for file in files_list:
    try:
        stock_data = pd.read_csv('normalised_data/'+file, index_col=0)
        s_i = stock_data.loc[month][features]
        r_i = stock_data.loc[month]['R']
        state_i.append(s_i)
        reward_i.append(r_i)
    except:
        continue
```

```python
def get_state_reward(time):
    i = time
    state_i = []
    reward_i = []
    month = months[i]
    for file in files_list:
        try:
            stock_data = pd.read_csv('normalised_data/'+file, index_col=0)
            s_i = stock_data.loc[month][features]
            r_i = stock_data.loc[month]['R']
            state_i.append(s_i)
            reward_i.append(r_i)
        except:
            continue
    return state_i, reward_i
```

```python
i = 100
state_i, reward_i = get_state_reward(100)
```
