```python
import pandas as pd
```

```python
data = pd.read_csv('DHAIRYA_IE620_Dataset.csv', low_memory=False)
```

```python
total_tickers = data.columns.unique()[7 + 6:3513:7]
```

```python
for tick in total_tickers:
    temp_data = data.loc[:, f'{tick}' : f'{tick}.6']
    temp_data[f'{tick}.6'] = temp_data[f'{tick}.6'].fillna(0)
    temp_data.columns = [temp_data.columns[0], 'price', 'volume', 'mc', 'pe', 'bm', 'div']
    temp_data.set_index(tick, inplace=True, drop=True)
    temp_data = temp_data[~temp_data.index.astype(str).str.fullmatch(r'\d+')]
    temp_data.index = pd.to_datetime(temp_data.index, errors='coerce')
    temp_data = temp_data.loc[temp_data.index.dropna(), :]
    temp_data = temp_data.dropna(subset=['price'])
    temp_data.to_csv(f'dataset/{tick}.csv')
```
